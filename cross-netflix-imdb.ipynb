```python
import re

def extract_re(pattern, array):
    return re.search(pattern, array).group(1)

TITLE_PATTERN = 'aria-label="(.*)" class="slider-refocus"'
```

```python

def extract_from_html(all_names_html):
    movie_names = []

    for name in all_names_html:
        try:
            movie_names.append(extract_re(TITLE_PATTERN, name))
        except:
            pass
    return movie_names
```

```python
TITLE_PATTERN = 'aria-label="(.*)" class="slider-refocus"'

def get_movie_names(html_path):
    print("get soup")
    soup = get_soup(html_path)
    print("Finding <a> tags")
    all_a = soup.findAll("a")
    print("Get all elements with aria-label subtag")
    all_names_html = [str(a) for a in all_a if 'aria-label' in str(a)]
    print("Extract name from html format")
    movie_names = extract_from_html(all_names_html)
       
    return movie_names
```

```python
TOP_250 = 'https://raw.githubusercontent.com/WittmannF/imdb-tv-ratings/master/top-250-movie-ratings.csv'
```

```python
top_movies = pd.read_csv(TOP_250, index_col=0)
```

```python
top_titles_set = set(top_movies['Title'].apply(lambda s: s.lower()).values)
```

```python
netflix_titles_set = df['Title'].apply(lambda s: s.lower()).values
```

```python
netflix_titles_set = set(netflix_titles_set)
```

```python
imdb_cross_netflix_set = top_titles_set & netflix_titles_set
```
